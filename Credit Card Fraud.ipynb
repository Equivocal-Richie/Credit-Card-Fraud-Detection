<p style="font-size:16px"><b>What is PCA Transformation?</b></p>
<p style="font-size:16px">Principal Component Analysis (PCA) reduces the number of dimensions in large datasets to principal components that retain most of the original information. It does this by transforming potentially correlated variables into a smaller set of variables, called principal components.</p>

<p style="font-size:16px"><b>Column Description:</b></p>

<ol style="font-size:16px">
    <li>
       `V1, V2, … V28` : These are the principal components obtained from the PCA. 
    </li><br>
    <li>
       `Time` : Contains the seconds elapsed between each transaction and the first transaction in the dataset
    </li><br>
    <li>
       `Amount` : Is the transaction Amount.
    </li><br>
    <li>
       'Class' : Is the response variable and it takes value 1 in case of fraud and 0 otherwise.
    </li>

<h1><b>3. Data Visualization</b></h1>

```python
# Let's see the number of fradulent cases vs normal transactions
fraud = credit_data[credit_data['Class'] == 1]
valid = credit_data[credit_data['Class'] == 0]
print("Fraud cases in our dataset: {}".format(len(credit_data[credit_data['Class'] == 1 ])))
print("Valid Transaction cases in our dataset: {}".format(len(credit_data[credit_data['Class'] == 0 ])))
```

```python
# Let's check the outlier fraction of inbalance in our dataset (target variable - 'Class')
outlier = len(fraud)/float(len(valid))*100
print("The percentage is: ",outlier)
```

```python
validity = credit_data['Class'].value_counts().to_frame()
validity
```

```python
# Let's visualize the data
plt.figure(figsize=(8, 4))
sns.countplot(x='Class', data=credit_data, hue="Class")
plt.title("Distribution of Fraud(1) vs Valid(0) Transactions")
plt.xlabel("Class")
plt.ylabel("Distribution")
plt.show()
```
